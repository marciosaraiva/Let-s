```python
window = Tk()
window.title("Jogo da Forca")

imagens = [
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca5.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca6.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca7.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca8.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca9.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca10.png"),
    PhotoImage(file="/home/marciosaraiva/Desktop/Lets Code/Turma_Abril_Maio/Projeto_Forca/imagens/forca11.png")
]


    
imgLabel = Label(window)
imgLabel.grid(row=1, column=0, columnspan=3, padx=10, pady=40)
imgLabel.config(image=imagens[0])

lblPalavra = StringVar()
Label(window, textvariable=lblPalavra, font=("Consolas 24 bold")).grid(row=1, column=3, columnspan=6, padx=10)


def checar(letra, palavra_correta):
    global numero_tentativas
    if numero_tentativas<6:
        palavra_exibicao = list(lblPalavra.get())
        lista_letras = list(palavra_com_espacos)
        
        if palavra_com_espacos.count(letra)>0:
            for posicao in range(len(lista_letras)):
                if lista_letras[posicao]==letra:
                    palavra_exibicao[posicao]=letra
                lblPalavra.set("".join(palavra_exibicao))
                
                if lblPalavra.get()==palavra_com_espacos:
                    messagebox.showinfo("Mensagem", "VocÃª venceu!")
                    messagebox.showinfo("A frase era...", frase)
                    run()
                    
        else:        
            numero_tentativas+=1
            imgLabel.config(image=imagens[numero_tentativas])
            
            if numero_tentativas==6:
                messagebox.showwarning("Mensagem", "Game Over")
                messagebox.showinfo("A frase era...", frase)
                run()
    



n=0
for letra in ascii_uppercase:
    Button(window, text = letra, command= lambda letra = letra:checar(letra, palavra_correta), font=("Helvetica 18"), width=4).grid(row=2+n//9, column=n%9)
    n+=1
    
Button(window, text="Novo\njogo", command=lambda:run(), font=("Helvetica 10 bold")).grid(row=4, column=8, sticky="NSWE")



def inicializa_interface(palavra):
    global palavra_com_espacos
    global numero_tentativas
    numero_tentativas = 0
    imgLabel.config(image=imagens[0])
    palavra_com_espacos = " ".join(palavra)
    lblPalavra.set(" ".join("_"*len(palavra)))
    
    

def run():
    global palavra_correta
    global frase
    
    # escolher o tema com interface grafica
    # tema = entryTema.get()
    # forca = Logica_Jogo.Jogo(tema)
    
    forca = Logica_Jogo.Jogo()
    dica = forca.exibe_dica()
    palavra_correta = forca.palavra_correta
    frase = forca.frase
    
    messagebox.showinfo("Dica", dica)
    
    palavra = forca.palavra_limpa
    inicializa_interface(palavra)
    
run()
window.mainloop()
```
